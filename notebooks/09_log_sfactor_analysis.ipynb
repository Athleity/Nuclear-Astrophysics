```python
import numpy as np

# Example: ln(S-factor) values in MeV·barn
ln_sfactor_list = list(map(float, input("Enter ln(S-factor) values (MeV·barn) separated by spaces: ").split()))

print("\nConverted S-factor values (keV·barn):")
for ln_S in ln_sfactor_list:
    # Take anti-log to get S in MeV·barn
    S_mev_barn = np.exp(ln_S)

    # Convert MeV·barn → keV·barn (multiply by 1000)
    S_kev_barn = S_mev_barn * 1000

    print(f"{S_kev_barn:.6f}")
```
