```python
import numpy as np

# --- Function to calculate S-factor error ---
def s_factor_error(S_factor, sigma, delta_sigma):
    """
    Computes error in S-factor using the relation:
    ΔS = (Δσ / σ) * S
    Parameters:
        S_factor (float or array): S(E) in MeV·barn
        sigma (float or array): Cross-section in millibarns (mb)
        delta_sigma (float or array): Uncertainty in cross-section (mb)
    Returns:
        delta_S (float or array): Uncertainty in S-factor (MeV·barn)
    """
    sigma = np.array(sigma, dtype=float)
    delta_sigma = np.array(delta_sigma, dtype=float)
    S_factor = np.array(S_factor, dtype=float)

    # Avoid division by zero
    with np.errstate(divide='ignore', invalid='ignore'):
        delta_S = np.where(sigma != 0, (delta_sigma / sigma) * S_factor, 0.0)

    return delta_S


# --- Example: Manual input from user ---
S_factor = np.array(list(map(float, input("Enter S-factor values (MeV·barn): ").split())))
sigma = np.array(list(map(float, input("Enter cross-section values (mb): ").split())))
delta_sigma = np.array(list(map(float, input("Enter delta cross-section values (mb): ").split())))

# Compute error in S-factor
delta_S = s_factor_error(S_factor, sigma, delta_sigma)

# --- Display results ---
print("\nEnergy (MeV)   σ (mb)   Δσ (mb)   S (MeV·barn)   ΔS (MeV·barn)")
print("-"*70)
for s, sig, dsig, dS in zip(S_factor, sigma, delta_sigma, delta_S):
    print(f"      -       {sig:8.3e}   {dsig:8.3e}   {s:12.5e}   {dS:12.5e}")
```
