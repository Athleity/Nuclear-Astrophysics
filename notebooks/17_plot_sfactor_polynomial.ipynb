```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# User Input
# -----------------------------
E = np.array(list(map(float, input("Enter Energy values (MeV): ").split())))
S_factor_keV = np.array(list(map(float, input("Enter S-factor values (keV·barn): ").split())))

# Convert S-factor from keV·barn → MeV·barn
S_factor = S_factor_keV / 1000.0

# -----------------------------
# Plotting
# -----------------------------
plt.figure(figsize=(8, 5))

# Line connecting actual data points only
plt.plot(E, S_factor, color='darkred', linewidth=2, marker='o',
         markersize=5, markerfacecolor='white', markeredgecolor='darkred',
         label='S-factor (calculated)')

# Labels and title
plt.title("Astrophysical S-Factor (via Polynomial) vs Energy", fontsize=14, fontweight='bold')
plt.xlabel("Energy (MeV)", fontsize=12)
plt.ylabel("S-Factor (MeV·barn)", fontsize=12)

# Grid and layout
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
plt.legend(fontsize=11, loc='best', frameon=True)
plt.tight_layout()

# Save and show
plt.savefig("S_factor_vs_energy_clean.png", dpi=400, bbox_inches='tight')
plt.show()
```
