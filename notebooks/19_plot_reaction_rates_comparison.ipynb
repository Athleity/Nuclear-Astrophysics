```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------
# User Input
# -------------------------------------
# Temperature values (T9 in GK)
T9 = np.array(list(map(float, input("Enter T9 (GK) values: ").split())))

# Reaction rates from integration (cm³ mol⁻¹ s⁻¹)
rate_computed = np.array(list(map(float, input("Enter Computed Reaction Rates: ").split())))

# Reaction rates from anti-log (cm³ mol⁻¹ s⁻¹)
rate_antilog = np.array(list(map(float, input("Enter Reaction Rates from Anti-log: ").split())))

# -------------------------------------
# Plotting
# -------------------------------------
plt.figure(figsize=(8, 5))

# Computed Reaction Rate
plt.plot(T9, rate_computed, color='#1f77b4', linewidth=2.2, marker='o',
         markersize=5, markerfacecolor='white', markeredgewidth=1.2,
         label='Computed (Integration)')

# Anti-log Reaction Rate — smoother, more visible
plt.plot(T9, rate_antilog, color='#d62728', linewidth=2.2, linestyle='--', marker='^',
         markersize=6, markerfacecolor='none', markeredgewidth=1.2,
         label='From Anti-log (Reference)')

# Labels and title
plt.title("Reaction Rate vs Temperature (T₉)", fontsize=15, fontweight='bold')
plt.xlabel("Temperature T₉ (GK)", fontsize=12)
plt.ylabel("Reaction Rate  (cm³ mol⁻¹ s⁻¹)", fontsize=12)

# Log scale (rates vary by many orders)
plt.yscale('log')

# Grid, legend, layout
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
plt.legend(fontsize=11, loc='best', frameon=True)
plt.tight_layout()

# Save and show
plt.savefig("ReactionRate_vs_T9_Clean.png", dpi=400, bbox_inches='tight')
plt.show()
```
