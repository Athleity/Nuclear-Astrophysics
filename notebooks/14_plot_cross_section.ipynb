```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# User Input Section
# -----------------------------
E = np.array(list(map(float, input("Enter Energy values (MeV): ").split())))
sigma = np.array(list(map(float, input("Enter Cross-section values (mb): ").split())))

# -----------------------------
# Plot Configuration
# -----------------------------
plt.figure(figsize=(7, 5))
plt.plot(E, sigma, color='navy', linewidth=2, marker='o', markersize=5, label='Experimental Data')

plt.title("Nuclear Reaction Cross-Section vs Energy", fontsize=14, fontweight='bold')
plt.xlabel("Energy (MeV)", fontsize=13)
plt.ylabel("Cross-Section (mb)", fontsize=13)

plt.grid(True, linestyle='--', linewidth=0.6, alpha=0.7)
plt.legend(fontsize=12, loc='best')
plt.tight_layout()

# -----------------------------
# Save & Show Plot
# -----------------------------
plt.savefig("cross_section_vs_energy.png", dpi=300, bbox_inches='tight')
plt.show()
```
