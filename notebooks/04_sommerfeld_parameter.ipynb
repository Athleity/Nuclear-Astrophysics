```python
import numpy as np

# Constants
amu_to_MeV = 931.49410242
e_squared = 1.44           # MeV·fm
hbar_c = 197.327           # MeV·fm

```

```python
# Charges
Z1 = 1  # Proton
Z2 = int(input("Enter atomic number (Z) of target nucleus (e.g., 21 for Sc): "))

# Masses
mass_projectile = 1.007276  # amu
mass_target = float(input("Enter mass of target nucleus (amu): "))
mass_product = float(input("Enter mass of residual nucleus (amu): "))

# Convert to MeV
M_proj = mass_projectile * amu_to_MeV
M_target = mass_target * amu_to_MeV
M_product = mass_product * amu_to_MeV

# Reduced mass
mu = (M_proj * M_target) / (M_proj + M_target)
```

```python
def lab_to_cm(E_lab, M_proj, M_target):
    return E_lab * M_target / (M_proj + M_target)

def compute_eta(Z1, Z2, mu, E_cm):
    return (Z1 * Z2 * e_squared / hbar_c) * np.sqrt(mu / (2 * E_cm))

def compute_S_factor(E_lab_list, sigma_list):
    results = []
    for E_lab, sigma in zip(E_lab_list, sigma_list):
        E_cm = lab_to_cm(E_lab, M_proj, M_target)
        eta = compute_eta(Z1, Z2, mu, E_cm)
        S = sigma * E_cm * np.exp(2 * np.pi * eta)
        results.append((E_lab, E_cm, sigma,eta, S / 1000))  # in MeV·barn
    return results
```

```python
# Energy values
E_input = input("Enter lab energies in MeV, separated by spaces: ")
E_lab_list = list(map(float, E_input.strip().split()))



# Cross-section values
σ_input = input("Enter corresponding cross sections in mb, separated by spaces: ")
sigma_list = list(map(float, σ_input.strip().split()))

if len(E_lab_list) != len(sigma_list):
    raise ValueError("Energy and cross-section lists must have the same length.")
```

```python
# Compute
results = compute_S_factor(E_lab_list, sigma_list)

# Output
print(f"\n{'E_lab (MeV)':>12} {'E_cm (MeV)':>12} {'σ (mb)':>10} {'eta':>10} {'S-factor (MeV·b)':>20}")
for E_lab, E_cm, sigma, eta, S in results:
    print(f"{E_lab:15.5f} {E_cm:15.5f} {sigma:12.5f} {eta:12.5f} {S:20.6f}")

```

```python
print("\neta:")
for _, _, _,eta, _ in results:
    print(f"{eta:.5f}")
```

```python
print("\ns_factor:")
for _, _, _,_,S in results:
    print(f"{S:.5f}")
```
