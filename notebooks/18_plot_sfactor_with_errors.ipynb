```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# User Input
# -----------------------------
E = np.array(list(map(float, input("Enter Energy values (MeV): ").split())))
S_factor = np.array(list(map(float, input("Enter S-factor values (MeV·barn): ").split())))
delta_S = np.array(list(map(float, input("Enter S-factor errors (ΔS in MeV·barn): ").split())))

# -----------------------------
# Plotting
# -----------------------------
plt.figure(figsize=(8, 5))

# Line connecting points
plt.plot(
    E, S_factor,
    color='darkred', linewidth=2,
    marker='o', markersize=5,
    markerfacecolor='white', markeredgecolor='darkred',
    label='S-factor (calculated)'
)

# Error bars (drawn after line for clarity)
plt.errorbar(
    E, S_factor, yerr=delta_S,
    fmt='none', ecolor='black', elinewidth=1.1,
    capsize=3, capthick=1
)

# Labels and title
plt.title("Astrophysical S-Factor vs Energy", fontsize=14, fontweight='bold')
plt.xlabel("Energy (MeV)", fontsize=12)
plt.ylabel("S-Factor (MeV·barn)", fontsize=12)

# Grid and layout
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
plt.legend(fontsize=11, loc='best', frameon=True)
plt.tight_layout()

# Save and show
plt.savefig("S_factor_vs_energy_with_error_line.png", dpi=400, bbox_inches='tight')
plt.show()

```
