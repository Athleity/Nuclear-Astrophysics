```python
import numpy as np

# === Physical constants ===
amu_to_MeV = 931.49410242   # 1 amu = MeV/c^2
e_squared = 1.44            # MeV·fm
hbar_c = 197.327            # MeV·fm

# === Projectile & Target ===
Z1 = 2                      # alpha particle charge
Z2 = 21                     # target charge (Scandium)
mass_projectile = 4.00260325415  # alpha in amu
mass_target = 44.95590828        # target in amu

# Convert to MeV
M_proj = mass_projectile * amu_to_MeV
M_target = mass_target * amu_to_MeV

# Reduced mass (in MeV/c^2)
mu = (M_proj * M_target) / (M_proj + M_target)

def lab_to_cm(E_lab, M_proj, M_target):
    return E_lab * M_target / (M_proj + M_target)

def compute_eta(Z1, Z2, mu, E_cm):
    return (Z1 * Z2 * e_squared / hbar_c) * np.sqrt(mu / (2 * E_cm))

# === Energy grid ===
Emax = 9.922
n = 96
E_lab = np.linspace(1e-6, Emax, n)
E_cm = lab_to_cm(E_lab, M_proj, M_target)

# === Polynomial fit for ln(S) ===
lnS = 0.0062*E_cm**3 - 0.2075*E_cm**2 + 1.0183*E_cm + 30.7
S = np.exp(lnS)  # S(E) in MeV·barn

# === Gamow factor ===
etas = compute_eta(Z1, Z2, mu, E_cm)
two_pi_eta = 2 * np.pi * etas
exp_minus_2pi_eta = np.exp(-two_pi_eta)

# === Compute σ(E) and Eσ(E) ===
sigma = (S / E_cm) * exp_minus_2pi_eta   # σ(E) in barns
E_sigma_E = E_cm * sigma                 # in MeV·barn

# === Print results ===
print(f"{'E_lab (MeV)':>12} {'E_cm (MeV)':>12} {'η':>12} {'2πη':>12} {'lnS':>12} {'S-factor (MeV·barn)':>20} {'Eσ(E) (MeV·barn)':>20}")
for e_lab, e_cm, eta, twopi, lnS_val, S_val, val in zip(E_lab, E_cm, etas, two_pi_eta, lnS, S, E_sigma_E):
    print(f"{e_lab:12.3f} {e_cm:12.3f} {eta:12.5f} {twopi:12.5f} {lnS_val:12.5f} {S_val:20.5e} {val:20.5e}")
```
