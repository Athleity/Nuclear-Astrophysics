```python
import matplotlib.pyplot as plt

# ====== USER INPUT SECTION ======
# Enter energy values (in MeV)
E = list(map(float, input("Enter energy values (MeV), separated by spaces:\n").split()))

# Enter S-factor values (from formula, in MeV·mb)
S_formula = list(map(float, input("Enter S-factor (formula-based) values, separated by spaces:\n").split()))

# Enter S-factor values (from polynomial fit, in MeV·mb)
S_poly = list(map(float, input("Enter S-factor (polynomial-based) values, separated by spaces:\n").split()))

# ====== PLOT SECTION ======
plt.figure(figsize=(8, 5))
plt.plot(E, S_formula, 'o-', label='S(E) from Formula', linewidth=2, markersize=6)
plt.plot(E, S_poly, 's--', label='S(E) from Polynomial Fit', linewidth=2, markersize=6)

# Labels and styling
plt.title("Astrophysical S-Factor vs Energy", fontsize=14)
plt.xlabel("Energy (MeV)", fontsize=12)
plt.ylabel("S-Factor (MeV·mb)", fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Show the plot
plt.show()
```
