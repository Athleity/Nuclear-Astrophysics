```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# User Input Section
# -----------------------------
E = np.array(list(map(float, input("Enter Energy values (MeV): ").split())))
sigma = np.array(list(map(float, input("Enter Cross-section values (mb): ").split())))
delta_sigma = np.array(list(map(float, input("Enter Cross-section errors (Δσ in mb): ").split())))

# -----------------------------
# Plot Configuration
# -----------------------------
plt.figure(figsize=(8, 5))

# Error bar plot with improved visibility
plt.errorbar(
    E, sigma, yerr=delta_sigma,
    fmt='o', markersize=5, markeredgewidth=1.2, markeredgecolor='black',
    color='navy', ecolor='crimson', elinewidth=1.4, capsize=5, capthick=1.2,
    label='Experimental Data (σ ± Δσ)'
)

# Optional connecting line
plt.plot(E, sigma, color='darkorange', linewidth=1.5, alpha=0.7, label='Trend Line')

# -----------------------------
# Styling and Labels
# -----------------------------
plt.title("Cross-Section vs Energy with Error Bars", fontsize=14, fontweight='bold')
plt.xlabel("Energy (MeV)", fontsize=13)
plt.ylabel("Cross-Section (mb)", fontsize=13)
plt.grid(True, linestyle='--', linewidth=0.6, alpha=0.7)
plt.legend(fontsize=11, loc='best', frameon=True)
plt.tight_layout()

# -----------------------------
# Save & Show Plot
# -----------------------------
plt.savefig("cross_section_error_visible.png", dpi=300, bbox_inches='tight')
plt.show()
```
