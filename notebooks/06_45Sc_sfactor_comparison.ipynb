```python
import numpy as np

def ln_S_next(E_lab):
    """Compute ln(S-factor) using the given cubic polynomial fit."""
    return (0.0153 * E_lab**3 - 0.4515 * E_lab**2 +
            3.3053 * E_lab + 28.436)

# Input energies
E_input = input("Enter lab energies in MeV, separated by spaces: ")
E_lab_list = list(map(float, E_input.strip().split()))

# Output only ln(S)
for E in E_lab_list:
    print(f"{ln_S_next(E):.6f}")
```
