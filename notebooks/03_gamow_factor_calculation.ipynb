```python
import numpy as np

# Constants
amu_MeV = 931.49410242  # 1 atomic mass unit in MeV/c^2
hbarc = 197.3269804     # ħ·c in MeV·fm
alpha = 1 / 137.035999084  # Fine-structure constant
e_sq = 1.4399764        # e² in MeV·fm
m_p = 938.27208816      # Proton mass in MeV/c^2
m_n = 939.56542052      # Neutron mass in MeV/c^2
```

```python
# Atomic masses in amu
masses = {
    '1H': 1.00782503223,
    'n': 1.00866491595,
    '45Sc': 44.95590828,
    '45Ti': 44.95812562,
    '48Ti': 47.94794198,
    '48V': 47.9522535,
    '51V': 50.94395704,
    '51Cr': 50.94820869,
    '52Cr': 51.94050623,
    '52Mn': 51.9455639,
    '55Mn': 54.93804391,
    '55Fe': 54.94229166,
    '56Fe': 55.93493633,
    '56Co': 55.9398388,
    '59Co': 58.93319429,
    '59Ni': 58.9343462,
    '62Ni': 61.92834488,
    '62Cu': 61.9340583,
    '63Cu': 62.92959772,
    '63Zn': 62.9332116,
    '66Zn': 65.92603381,
    '66Ga': 65.9318084
}
```

```python
def reduced_mass(m1, m2):
    """Reduced mass in MeV/c^2 and in amu"""
    mu_MeV = (m1 * m2) / (m1 + m2)
    mu_amu = mu_MeV / amu_MeV
    return mu_MeV, mu_amu

def calculate_q_value(target, product):
    """Q-value for (p,n) reaction in MeV"""
    M_target = masses[target] * amu_MeV
    M_product = masses[product] * amu_MeV
    return (M_target + m_p) - (M_product + m_n)

def calculate_threshold(Q, M_target):
    """Threshold energy in LAB and CM frame if Q < 0"""
    if Q >= 0:
        return 0, 0
    E_cm = -Q
    E_lab = -Q * (M_target + m_p) / M_target
    return E_lab, E_cm

def gamow_factor(Z1, Z2, mu, E_cm):
    """Returns the Gamow factor G(E) = exp(-2π * η)"""
    if E_cm <= 0:
        return np.nan
    k = np.sqrt(2 * mu * E_cm) / hbarc  # wave number
    eta = (Z1 * Z2 * e_sq * mu) / (hbarc * hbarc * k)
    return np.exp(-2 * np.pi * eta)
```

```python
reactions = [
    ('45Sc', '45Ti'), ('48Ti', '48V'), ('51V', '51Cr'),
    ('52Cr', '52Mn'), ('55Mn', '55Fe'), ('56Fe', '56Co'),
    ('59Co', '59Ni'), ('62Ni', '62Cu'), ('63Cu', '63Zn'),
    ('66Zn', '66Ga')
]

# Atomic number map
Z_map = {
    'Sc': 21, 'Ti': 22, 'V': 23, 'Cr': 24, 'Mn': 25,
    'Fe': 26, 'Co': 27, 'Ni': 28, 'Cu': 29, 'Zn': 30
}
```

```python
print("{:<15} {:>8} {:>10} {:>10} {:>10} {:>12}".format(
    "Reaction", "Q (MeV)", "E_lab", "E_cm", "μ (amu)", "Gamow Factor"
))

for target, product in reactions:
    A1 = int(''.join(filter(str.isdigit, target)))  # Mass number
    element = ''.join(filter(str.isalpha, target))  # Element symbol
    Z1 = Z_map[element]
    Z2 = 1  # Proton

    Q = calculate_q_value(target, product)
    M_target = masses[target] * amu_MeV
    E_lab, E_cm = calculate_threshold(Q, M_target)
    mu_MeV, mu_amu = reduced_mass(m_p, M_target)
    G = gamow_factor(Z1, Z2, mu_MeV, E_cm)

    print("{:<15} {:>8.3f} {:>10.3f} {:>10.3f}  {:>10.5f} {:>12.2e}".format(
        f"{target}(p,n){product}", Q, E_lab, E_cm, mu_amu, G
    ))
```
